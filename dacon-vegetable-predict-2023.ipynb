```python
train['item_id'] = train['ID'].str[:6]

# 날짜 변환
train['timestamp'] = pd.to_datetime(train['timestamp'])
test['timestamp'] = pd.to_datetime(test['timestamp'])

train['year'] = train['timestamp'].dt.year # 년도
train['month'] = train['timestamp'].dt.month # 월
train['day'] = train['timestamp'].dt.day # 일

test['year'] = test['timestamp'].dt.year # 년도
test['month'] = test['timestamp'].dt.month # 월
test['day'] = test['timestamp'].dt.day # 일


train['weekday'] = train['timestamp'].dt.weekday
test['weekday'] = test['timestamp'].dt.weekday

# Quarter 변수 생성
train['quarter'] = 0 # 겨울
train.loc[(train['month']>=3) & (train['month']<=5),'quarter']=1 # 봄
train.loc[(train['month']>=6) & (train['month']<=8),'quarter']=2 # 여름
train.loc[(train['month']>=9) & (train['month']<=11),'quarter']=3 # 가을

test['quarter'] = 0 # 겨울
test.loc[(test['month']>=3) & (test['month']<=5),'quarter']=1 # 봄
test.loc[(test['month']>=6) & (test['month']<=8),'quarter']=2 # 여름
test.loc[(test['month']>=9) & (test['month']<=11),'quarter']=3 # 가을
```

명절/휴일 데이터 추가

```python
import holidays
holiday_list = train.timestamp
holiday_list
```

```python
year_list = [2019,2020,2021,2022,2023]
kr_holiday = []
kr_myeongjul = []
myeongjul = ["Lunar New Year's Day","Chuseok"]
for arg in year_list:
    # 한국 휴일 객체 생성 
    kr_holidays = holidays.KR(arg)
    di = dict(kr_holidays)
    kr_holidays = {v:k for k,v in di.items()}
    
    tt=list(kr_holidays.keys())
    tt.remove("Lunar New Year's Day")
    tt.remove("Chuseok")
    hol = tt
    
    for arg_1 in myeongjul:
        kr_myeongjul.append(kr_holidays[arg_1])
    for arg_2 in hol:
        kr_holiday.append(kr_holidays[arg_2])
    
kr_myeongjul
```

```python
# generate holiday table
train['holiday'] = train.timestamp.apply(lambda x: 1 if x.date() in kr_myeongjul else (2 if x.date() in kr_holiday else 0))
```

```python
timeseries_data = TimeSeriesDataFrame(train.drop(columns=['ID','기간','corporation','location','item','Date']))
```

```python
timeseries_data
```

```python
model = TimeSeriesPredictor(prediction_length = 28,target = 'price(원/kg)',eval_metric='RMSE')
model
```

```python
model.fit(timeseries_data,random_seed = 42)
```

```python
# -712.6053?, leaderboard : 686
```

```python
result = model.predict(timeseries_data,random_seed=42)
```
